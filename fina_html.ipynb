```python
import requests
import re
from bs4 import BeautifulSoup
import csv
url = 'https://nuggetsnetwork.com/blog/Data-Science-Jobs-In-India.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
soup1 = str(soup)
soup2 = soup1.split('\n')
job = []
company = []
location = []
urls = []
j = 1
c = 1
l = 1
u = 1
with open('/content/linkedin3.csv', 'r', encoding='latin1') as exl:
  exl1 = csv.reader(exl)
  for i in exl1:
    job.append(i[0].strip(' '))
    company.append(i[1].strip(' '))
    location.append(i[2].strip(' '))
    urls.append(i[3])
#print(location)

#print(soup2)
modified = []
for i in soup2:
    modified.append(i)
#    print(i)
    if re.search('^<h3>',i):
        modified.remove(i)
        modified.append('<h3>'+job[j]+'</h3>')
        j = j +1
    elif re.search('^<p><b>Location:',i):
      modified.remove(i)
      modified.append('<p><b>Location:</b>'+location[l]+'<br/><b>Company:</b>'+company[c]+'<br/></p>')
      l = l+1
      c = c+1
    elif re.search('^<a class="btn btn-outline-dark',i):
      modified.remove(i)
      modified.append(urls[u])
      u = u+1
#    modified.append(i)
new_html = []
for n in modified:
  new_html.append(n)
final_html = '\n'.join(new_html)
html_file = open('final.html', 'w')
html_file.write(final_html)
print(final_html)


      
#  modified.append(i)
#print(modified)
#writer = open('soup.txt', 'w')
#write1 = writer.write(soup1) 

#html1 = open('html.txt', 'w')







```
